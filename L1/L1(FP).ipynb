```python
act = 0
if len(a) <= 0:
    print("Помилка. Список пустий або присутнє тільки одне число")
    check = False
        
for i in range(len(a)):
    try:    
        int(a[i])
    except ValueError:
        print('Помилка. У списку присутнє нечислове значення')
        check = False

if check == True: 
    for i in range(len(a)-1):
        for j in range(len(a)-i-1):
            if a[j] > a[j+1]:
                a[j], a[j+1] = a[j+1], a[j]
                act+=1
if check == True: 
    print("Відсортований список:", a)
    print("Кількість перестановок:", act)
```
