### Вправа №2

Вводиться число n. Розставити 1 та 0 у шаховому порядку,
починаючи з 1 в матриці розміром nхn, використовуючи слайсінг.

                                             Рисунок 1 - Приклад роботи програми до вправи №2

#### Код програми:

```python
n = 6
```

```python
a = np.zeros((n, n), dtype = 'int16')
a[1::2,::2] = 1
a[::2,1::2] = 1
```

```python
print(a)
```

### Вправа №8

З клавіатури вводиться масив. Підрахувати в ньому кількість
нульових та ненульвих елементів.

                                             Рисунок 3 - Приклад роботи програми до вправи №8

#### Код програми:

```python
c = [3, 4, 0, 9, 8, 2, 4, 0, 8, 4, 0]
c = np.array(c)
```

```python
c1 = c.copy()
c1 = c1[c1==0]
c1.size
```

```python
c1 = c.copy()
c1 = c1[c1>0]
c1.size
```
