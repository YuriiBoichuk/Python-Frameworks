### Вправа №18

Заповнити вектор значеннями від 0 до n. Замінити знаки для всіх
значеннь, що менші за n/2 та більші за 3n/4 на протилежні

                                             Рисунок 8 - Приклад роботи програми до вправи №18

#### Код програми:

```python
n = 10
```

```python
h = np.arange(n+1)
l, r = np.split(h, [n//4+1])
l[l<n/2] *= -1
r[r > 3*n/4] *= -1
print(h)
```

## Завдання №2

Розв’язати систему за допомогою формул Крамера і виконати перевірку за
допомогою:
 - матричного множення;
 - функції numpy.linalg.solve().

                         Рисунок 10 - Система лінійних алгебраїчних рівнянь до виконання завдання №2

#### Код програми:

```python
a = np.array([[1, -3, 4, -5],
            [1, -2, 3, -4],
            [0, 2, -5, 12],
            [4, 3, -5, 5]])
b = np.array([3, 6, 4, -5])
```

```python
detA = np.linalg.det(a)
print('Детермінант основної матриці =', detA)
x = a.copy()
x[::,0] = b
print('\n{}\n\nДетермінант = {}'.format(x, np.linalg.det(x)))
x1 = np.linalg.det(x) / detA
x = a.copy()
x[::,1] = b
print('\n{}\n\nДетермінант = {}'.format(x, np.linalg.det(x)))
x2 = np.linalg.det(x) / detA
x = a.copy()
x[::,2] = b
print('\n{}\n\nДетермінант = {}'.format(x, np.linalg.det(x)))
x3 = np.linalg.det(x) / detA
x = a.copy()
x[::,3] = b
print('\n{}\n\nДетермінант = {}'.format(x, np.linalg.det(x)))
x4 = np.linalg.det(x) / detA
print('\nВласна реалізація методу Крамера:\n', x1, '\n', x2, '\n', x3, '\n', x4)
print('Перевірка методом linalg.solve():', np.allclose(np.linalg.solve(a, b), [x1, x2, x3, x4]))
```

## Завдання №4

Розв’язати рівняння за допомогою оберненої матриці і виконати перевірку
за допомогою:
- матричного множення;
- функції ``numpy.linalg.solve().``

                                    Рисунок 12 - Матричне рівняння до виконання завдання №4

#### Код програми:

```python
a = np.array([[2, 1, 3],
            [1, -2, 0],
            [4, -3, 0]])
b = np.array([[22, -14, 3],
            [6, -7, -0],
            [11, 3, 15]])
x = np.linalg.inv(a)@b
print('Множенням матриць A^(-1)*B отримали X:\n', x)
print('Перевірка, A*X:\n', np.allclose(a@x, b))
```
